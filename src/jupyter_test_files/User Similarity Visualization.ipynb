```python
import networkx as nx
G = nx.Graph()
import matplotlib.pyplot as plt
options = {
     'node_color': 'black',
     'node_size': 100,
     'width': 3,
}
G.add_node(0)
i=1
G.add_node(0,pos=(6,1))
```

```python
A = [(856, 1.0), (515, 1.0), (2, 1.0), (725, 1.0), (957, 0.99647864852674928), (433, 0.99498743710661997), (36, 0.99444479713328215)]
```

```python
A
```

```python
for each in A:
    i +=1
    G.add_node(each[0],pos = (i,i))
    G.add_edge(0,each[0],weight=each[1])
pos=nx.get_node_attributes(G,'pos')

labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
nx.draw(G,pos)
plt.show()
```
