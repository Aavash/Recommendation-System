# Data Visualization with Pandas and matplotlib


```python
from copy import deepcopy
import pandas as pd
initial_data = pd.read_csv('data.csv')
initial_data.head()
import numpy as np
```

```python
data = deepcopy(initial_data)
data.set_index('Id', inplace=True)
```

```python
data.head()
```

```python
initial_data.head()
```

```python
data['php'] = [1 if i=='PHP' else 0 for i in data['Interest']]
data['python'] = [1 if i=='python' else 0 for i in data['Interest']]
data['qa'] = [1 if i=='QA' else 0 for i in data['Interest']]
data['js'] = [1 if i=='JS' else 0 for i in data['Interest']]
data.head()
```

```python
data.drop('Interest', axis=1, inplace=True)
data.head()
```

```python
data['qualification'] = [1 if v == '+2' else 2 if v == 'Bachelors' else 3 for v in data['Qualification']]
# ['yes' if v == '+2' else 'no' if v == 'Bachelors' else 'idle' for v in data['Qualification']]
```

```python
data.head()
```

```python
data['level'] = [1 if v == 'Entry' else 2 if v == 'Mid' else 3 for v in data['Level']]
```

```python
data.drop(['Qualification', 'Level'], axis=1, inplace=True)
```
